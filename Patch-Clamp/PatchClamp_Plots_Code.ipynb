```python
## Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
import pyabf

homeDir = 'xxxxx'         
# Directory containing Python code

dataDir = 'xxxxx' 
# Directory containing input datafile
```

```python
## Load Data

os.chdir(dataDir)

abf = pyabf.ABF("2021_10_26_0041.abf") # Alter name of datafile of interest
os.chdir(homeDir)

os.getcwd()
```

```python
## Plot Maximum Hyperpolarization, First AP and AP Train

w, h = 6.5,4.5
ncols = 2
nrows = 2

plt.figure(figsize=(ncols*w, nrows*h))

abf.setSweep(0)
plt.plot(abf.sweepX, abf.sweepY, color='k')
abf.setSweep(9)
plt.plot(abf.sweepX, abf.sweepY, color='g')
abf.setSweep(11)
plt.plot(abf.sweepX, abf.sweepY, color='k')

plt.vlines(0.4,0,20,color='k',linestyles='solid',linewidths = 1.5) # Scale bar y
plt.ylim([-120,60])
plt.xlim([0.23,0.55])
plt.xlim([0.23,1.575])
plt.xlabel('Time (s)')
plt.ylabel('Membrane Potential (mV)')

plt.show()
```
